```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class ComplimentState(TypedDict):
    name: str


def compliment_node(state: ComplimentState) -> ComplimentState:
    state["name"] = f'{set["name"]} youre doing an amazing job learning LangGraph!'

    return state
```

```python
graph = StateGraph(ComplimentState)
graph.add_node("name", compliment_node)
graph.set_entry_point("name")
graph.set_finish_point("name")

app = graph.compile()

result = app.invoke({"name": "Bob"})

result["name"]
```
